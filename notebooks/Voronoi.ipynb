```python
import numpy as np
from scipy.spatial import Voronoi, voronoi_plot_2d 
import matplotlib.pyplot as plt
```

```python
points = np.array([[0, 0], [0, 1], [0, 2], [1, 0], [1, 1], [1, 2], [2, 0], [2, 1], [2, 2]])
```

```python
vor = Voronoi(points)
voronoi_plot_2d(vor)
plt.show()
```

```python
sites = np.random.rand(15,2)
# add 4 distant dummy points
sites = np.append(sites, [[1,1], [0,0], [0,1], [1,0]], axis = 0)
sites.shape
```

```python
# compute Voronoi tesselation
vor1 = Voronoi(sites)
# plot
voronoi_plot_2d(vor1)
```

```python
# colorize
for region in vor1.regions:
    if not -1 in region:
        polygon = [vor1.vertices[i] for i in region]
        plt.fill(*zip(*polygon))

# fix the range of axes
plt.xlim([0,1]), plt.ylim([0,1])

plt.show()
```

### Drawing Line between two points

(y-y1)/(x-x1) = (x2-x1)/(y2-y1)
y-y1 = m(x-x1)
=> y = mx - mx1+y1

```python
def line_joining_points(P,Q):
#     P(x1, y1),   Q(x2, y2)
    b = -1*(Q[0] - P[0])
    a = Q[1] - P[1]
    c = a*P[0] + b*P[1]
    mid_point = ((P[0]+Q[0])/2, (P[1]+Q[1])/2)
    
    if b<0:
        print('Equation of line is {}x {}y = {}' .format(a, b, c))
    else:
        print('Equation of line is {}x+ {}y = {}' .format(a,b,c))
     
    c = b*mid_point[0]-a*mid_point[1]
    temp = a
    a = b
    b = -1*temp
    
    if a<0:
        print('Equation of perpendicular bisector is {}x+ {}y = {}' .format(a, b, c))
    else:
        print('Equation of perpendicular bisector is {}x {}y = {}' .format(a, b, c))
```

```python
P = [3, 2] 
Q = [2, 6]  
line_joining_points(P, Q)
```

```python
def midPoint_and_slope(p1, p2):
    '''
    Finds the midpoint between two points and the slope of the perpendicular bisector
    
    Input:
        - p1: Tuple of point 1: (x,y)
        - p2: Tuple of point 2: (x,y)
    Output:
        - midPoint: Tuple (x,y)
        - slope: Float
    '''
    midPoint = ((p1[0]+ p2[0])/2, (p1[1] + p2[1])/2)
    
    slope_og = (p2[1] - p1[1])/(p2[0] - p1[0])
    
    slope = - 1/slope_og
    
    return midPoint, slope
```

## Plotting line and its perpendicular bisector

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_line(P,Q):
    mid_point = ((P[0]+Q[0])/2, (P[1]+Q[1])/2)
    
    m = ((Q[1]-P[1])/(Q[0]-P[0]))
    c1 = -1*m*P[0]+P[1]
    x = np.arange(1, 4, 0.1)
    y = m*x + c1
    m_normal = (-1/m)
    
    x1 = np.arange(1, 4, 0.1)
    c_normal = (-1*m_normal*mid_point[0]) + mid_point[1]
    y1 = m_normal*x1 + c_normal
    
    point_x = [P[0], Q[0], mid_point[0]]
    point_y = [P[1], Q[1], mid_point[1]]
    
    plt.scatter(point_x, point_y)
    plt.plot(x,y)
    plt.plot(x1,y1)
    return
```

```python
P = [3, 2] 
Q = [2, 6]  
plot_line(P, Q)
```
